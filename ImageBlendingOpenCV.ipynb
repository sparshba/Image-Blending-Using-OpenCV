```python
import numpy as np
import pandas as pd
import cv2 as cv  # import openc
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
def read_file(filename):
    img=cv.imread(filename)
    img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    return img
```

```python
filename="messi5.jpg"
img1=read_file(filename)
```

```python
print(img1.shape)
```

```python
print(img1[100,100])
```

```python
# add another ball to the image
ball = img1[280:340, 330:390]
img1[273:333, 100:160] = ball

plt.imshow(img1)
```

```python
filename="opencv.jpg"
img2=read_file(filename)
```

```python
print(img2.shape)
```

```python
img2a = cv.resize(img2, (548, 342)) # resizing opencv logo to be the same size as messi
print(img2a.shape)
plt.imshow(img2a)
```

```python
# Refer to documentation: https://docs.opencv.org/master/d2/de8/group__core__array.html#gafafb2513349db3bcff51f54ee5592a19
dst = cv.addWeighted(img1,0.7,img2a,0.3,0)
plt.imshow(dst)
```

```python
dst = cv.addWeighted(img1,0.3,img2a,0.7,100)
plt.imshow(dst)
```
